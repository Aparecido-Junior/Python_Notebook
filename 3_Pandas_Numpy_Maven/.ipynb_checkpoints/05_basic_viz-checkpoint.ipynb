```python
(df
 .query("store_nbr == 4")
 .set_index("date")
 .loc["2013-01", "transactions"]
 .plot()
);
```

```python
(df
 .query("store_nbr in [44, 47]")
 .pivot_table(index= "date", columns= "store_nbr")
 .loc["2013-01", :]
 .droplevel(0, axis=1)
 .plot()
);
```

## Title

```python
(df
 .query("store_nbr in [44, 47]")
 .pivot_table(index= "date", columns= "store_nbr")
 .loc["2013-01", :]
 .droplevel(0, axis=1)
 .plot(title="Comparison store 44 and 47 over Jan 2013 ", xlabel="Jan 2013", ylabel="Num transactions")
);
```

## Colors

```python
(df
 .query("store_nbr in [44, 47]")
 .pivot_table(index= "date", columns= "store_nbr")
 .loc["2013-01", :]
 .droplevel(0, axis=1)
 .plot(title="Comparison store 44 and 47 over Jan 2013 ", 
       xlabel="Jan 2013", 
       ylabel="Num transactions",
       color=["red","black"])
);
```

## Line Style

```python
(df
 .query("store_nbr in [44, 47]")
 .pivot_table(index= "date", columns= "store_nbr")
 .loc["2013-01", :]
 .droplevel(0, axis=1)
 .plot(title="Comparison store 44 and 47 over Jan 2013 ", 
       xlabel="Jan 2013", 
       ylabel="Num transactions",
       style=['-','--'])
);
```

## Chart Legend and Grid

```python
(df
 .query("store_nbr in [44, 47]")
 .pivot_table(index= "date", columns= "store_nbr")
 .loc["2013-01", :]
 .droplevel(0, axis=1)
 .plot(title="Comparison store 44 and 47 over Jan 2013 ", 
       xlabel="Jan 2013", 
       ylabel="Num transactions",
       style=['-','--'])
).legend(bbox_to_anchor=(1, 1));
```
