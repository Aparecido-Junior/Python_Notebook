### Numpy Select

```python
oil = ( 
    pd.read_csv(r'C:\Users\User\OneDrive\Documentos\Data_Analytcs\Python\Pandas+Course+Resources\Pandas\retail\oil.csv')
)

oil.head()
```

```python
oil.columns = ['date','price']

oil.head()
```

```python
oil['buy'] = np.where(oil['price']>100, 'too high', 'buy')

oil
```

```python
conditions = [
    (oil['price'] > 100),
    (oil['price'] <=100) & (oil['price'] > 50),
    (oil['price'] <50)
]

choices = ['dont buy','buy','strong buy']

oil['buy'] = np.select(conditions, choices, default='missing')

oil.head()
```
