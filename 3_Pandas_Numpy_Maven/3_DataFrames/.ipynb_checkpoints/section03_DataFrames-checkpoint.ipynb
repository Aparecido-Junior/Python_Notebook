# Assignment 6: Filtering DataFrames (REFAZER)

I need some quick research on store 25:

* First, calculate the percentage of times ALL stores had more than 2000 transactions
* Then, calculate the percentage of times store 25 had more than 2000 transactions, and calculate the sum of transactions on these days
* Finally, sum the transactions for stores 25 and 31, that occurred in May or June, and had less than 2000 transactions


```python
transactions = pd.read_csv("transactions.csv", parse_dates=['date'])
transactions.info()
```

```python
# First, calculate the percentage of times ALL stores had more than 2000 transactions.
print((transactions["transactions"] > 2000).mean()) 
```

```python
# Then, calculate the percentage of times store 25 had more than 2000 transactions, and calculate the sum of transactions on these days.
mask = transactions.query("store_nbr == 25 and transactions > 2000")

print(mask['transactions'].sum())
```

```python
# Finally, sum the transactions for stores 25 and 31, that occurred in May or June, and had less than 2000 transactions.
mask = transactions.query("store_nbr == (25, 31) and date.dt.month == [5, 6] and transactions < 2000")

print(mask['transactions'].sum())
```

# Assignment 9: Column Creation

Just some quick work, but can you send me the transaction data with the columns renamed?

* Create a `pct_to_target` column that divides transactions by 2500.
* Then, create a `met_target` column that returns True if `pct_to_target` is greater than or equal to 1.
* Next, create a `bonus_payable` column that equals 100 if `met_target` is True, and 0 if not. Then sum the bonus payable column.
* Finally, create columns for month and day of week as integers. There is some helper code for these dateparts below.



Thanks!



```python
transactions = pd.read_csv('transactions.csv')

transactions.head()
```

```python
# Create a pct_to_target column that divides transactions by 2500.

transactions['pct_to_target'] = transactions.loc[:, 'transactions'] / 2500

transactions.head()
```

```python
# Then, create a met_target column that returns True if pct_to_target is greater than or equal to 1.

transactions['met_target'] = transactions.loc[:, 'pct_to_target'] >= 1

transactions.head()
```

```python
# Next, create a bonus_payable column that equals 100 if met_target is True, and 0 if not. Then sum the bonus payable column.
transactions['bonus_payble'] = np.where(transactions['met_target'] == True, 100, 0)

transactions.head()

print(transactions['bonus_payble'].sum())
```

```python
#Finally, create columns for month and day of week as integers. There is some helper code for these date parts below.
transactions['month'] = transactions['date'].str[5:7]

transactions.head()
```

# Assignment 10: np.select

Hi there! I need a few columns created.

1. Create a ‘seasonal_bonus’ column that applies to these dates: 
    * All days in December (month = 12)
    * Sundays (day_of_week = 6) in May (month = 5)
    * Mondays (day_of_week = 0) in July (month = 7)
2. Call the December bonus ‘Holiday Bonus’, the May bonus ‘Corporate Month’, and the July bonus ‘Summer Special’. If no bonus applies, the column should display ‘None’. 
3. Finally, calculate the total bonus owed at $100 per day.

Thanks!

```python
transactions = pd.read_csv('transactions.csv')

transactions.head()
```

```python
conditions = [
    (transactions['date'].str[5:7] == '12'),
    (transactions['date'].str[5:7] == '05'),
    (transactions['date'].str[5:7] == '07')
]

choices = ['Holiday Bonus', 'Corporate Month', 'Summer Special']

transactions['seasonal_bonus'] = np.select(conditions, choices, default='none')

transactions['seasonal_bonus'].value_counts()
```

```python
transactions = pd.read_csv(
    r"C:\Users\User\OneDrive\Documentos\Data_Analytcs\Python\Pandas_Course_Resources\Pandas\retail\transactions.csv")

transactions.date.str[-2:]
```

```python
'''
Create a pct_to_target column that divides transactions by 2500.
Then, create a met_target column that returns True if pct_to_target is greater than or equal to 1.
Next, create a bonus_payable column that equals 100 if met_target is True, and 0 if not. Then sum the bonus payable column.
Finally, create columns for month and day of week as integers. There is some helper code for these dateparts below.
'''
```

```python
# Create same columns with assign
transactions = transactions.assign(
    pct_to_target = transactions["transactions"] / 2500,
    met_target = lambda transactions: transactions["pct_to_target"] > 1,
    bonus_payble = lambda transactions: np.where(transactions["met_target"] == True, 100, 0),
    month = lambda transactions: transactions["date"].str[5:7],
    day = lambda transactions: transactions["date"].str[-2:]
)

transactions.head()
```

```python
'''
Create a ‘seasonal_bonus’ column that applies to these dates:
All days in December (month = 12)
Sundays (day_of_week = 6) in May (month = 5)
Mondays (day_of_week = 0) in July (month = 7)
Call the December bonus ‘Holiday Bonus’, the May bonus ‘Corporate Month’, and the July bonus ‘Summer Special’. If no bonus applies, the column should display ‘None’.
Finally, calculate the total bonus owed at $100 per day.
'''
```

```python
transactions = transactions.assign(
    seasonal_bonus = np.where(transactions["month"] == "12", "Holiday Bonus", 
    np.where((transactions["month"]== "05") & (transactions["day"] == "06"), "Corporate Month",
    np.where(transactions["month"]== "07" , "Summer Special", "None")
)))   
 
transactions.head()

transactions.head()
```
