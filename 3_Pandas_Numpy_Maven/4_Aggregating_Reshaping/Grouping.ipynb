```python
import pandas as pd
```

```python
df = pd.read_excel(r"..\project_data\premier_league_games_full.xlsx")
df.head()
```

```python
df.groupby("HomeTeam")["HomeGoals"].sum().head() #Series
df.groupby("HomeTeam")[["HomeGoals"]].sum().head() #Dataframe
df.groupby("HomeTeam")[["HomeGoals"]].sum().head().sort_values('HomeGoals', ascending=False)
```

```python
df.groupby(["HomeTeam", "season"])[["HomeGoals"]].sum()
df.groupby(["season", "HomeTeam"])[["HomeGoals"]].sum()
df.groupby(["season", "HomeTeam"], as_index=False)[["HomeGoals"]].sum()
df.groupby(["season", "HomeTeam"], as_index=False)[["HomeGoals"]].sum().query("HomeTeam == 'Arsenal'")
```

```python
df.groupby(["season", "HomeTeam"])[["HomeGoals"]].sum().head()
df.groupby(["season", "HomeTeam"])[["HomeGoals"]].sum().loc['2008/2009']
df.groupby(["season", "HomeTeam"])[["HomeGoals"]].sum().loc['2008/2009': "2010/2011"]
df.groupby(["season", "HomeTeam"])[["HomeGoals"]].sum().loc[('2008/2009', "Arsenal")]
df.groupby(["season", "HomeTeam"])[["HomeGoals"]].sum().loc[('2008/2009', "Arsenal"): ('2008/2009', "Bolton Wanderers")]
```

```python
df.groupby(["season", "HomeTeam"]).agg({"HomeGoals": ['sum', 'mean']})
df.groupby(["season", "HomeTeam"]).agg({"HomeGoals": ['sum', 'mean']}).iloc[1, :]
df.groupby(["season", "HomeTeam"]).agg({"HomeGoals": ['sum', 'mean']}).loc["2010/2011", ('HomeGoals', 'mean')].head()
```
