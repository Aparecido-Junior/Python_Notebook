## Pivot Tables

```python
(
    df.pivot_table(index="HomeTeam",
             columns="season",
             values="HomeGoals",
             aggfunc="sum")
)   

(
    df.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']")
      .pivot_table(index="HomeTeam",
             columns="season",
             values="HomeGoals",
             aggfunc="sum",
             margins=True)
)  
```

## Mult Aggregations Pivot Table

```python
(
    df.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']")
      .pivot_table(index="HomeTeam",
             columns="season",
             values="HomeGoals",
             aggfunc=["sum", "mean"],
             margins=True)
)  
```

## HeatMaps Pivot Table

```python
(
    df.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']")
      .pivot_table(index="HomeTeam",
             columns="season",
             values="HomeGoals",
             aggfunc="mean",
             margins=True)
      .style.background_gradient(cmap="RdYlGn", axis=1)
)  
```

## Melt

```python
(
    df.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']")
      .pivot_table(index="HomeTeam",
             columns="season",
             values="HomeGoals",
             aggfunc="mean")
    .reset_index()
    .melt(id_vars="HomeTeam",
         value_vars=["2008/2009", "2009/2010", "2010/2011"],
         value_name= "avg_goals")
    .style.background_gradient(cmap="RdYlGn", axis=None)
)  
```
