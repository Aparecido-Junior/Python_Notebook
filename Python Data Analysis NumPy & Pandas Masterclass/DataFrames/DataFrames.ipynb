### Filtering DataFrames



```python
oil = pd.read_csv('oil.csv', parse_dates=['date'])

oil.columns = ['Date','Price']

oil['benchmark'] = 100

oil.head()
```

```python
oil.describe()
```

```python
oil.loc[oil['Price'] > 100].head()
oil.loc[oil['Price'] > oil['benchmark']].head()
```

```python
oil.loc[oil['Date'].str[:4] == '2013'].head()
```

```python
mask = ((oil['Date'].str[:4] == '2013') 
        & (oil['Price'] > oil['benchmark']))
```

```python
oil.loc[mask].head()
```

### Query

```python
oil.query('Price > benchmark')
oil.query('Price > benchmark or Date.dt.year == 2013')
```

### Sorting

```python
oil = pd.read_csv('oil.csv', parse_dates=['date'])

oil.columns = ['Date','Price']

oil['benchmark'] = 100

oil.head()
```

```python
oil.sort_index(ascending=False)
oil.sort_index(ascending=True, axis=1)
```

```python
oil.sort_values('Price', ascending=False)
oil.sort_values(['Price', 'Date'], ascending=[False, True]).head()
```
