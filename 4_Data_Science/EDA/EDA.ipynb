```python
import pandas as pd
```

### Filtering

```python
groceries = pd.read_pickle(r'..\groceries_with_new_columns.pkl')
groceries.head()
```

```python
groceries[groceries['Low Inventory'] == 'Low Inventory'] #filtering the whole data by one condition
groceries.loc[groceries['Low Inventory'] == 'Low Inventory'] #filtering using .loc
groceries.loc[groceries['Low Inventory'] == 'Low Inventory', 'Price_Dollars'] # Filtering one column of the data, must use .loc
groceries.loc[groceries['Low Inventory'] == 'Low Inventory', ['Price_Dollars','Inventory']] # Filtering two or more columns of the data, must use .loc
```

```python
groceries.Subcategory = groceries.Subcategory.str.strip() # when filtering, must pay attention to text type

groceries[groceries.Subcategory == 'Snacks']
```

```python
mask = (groceries['Low Inventory'] == 'Low Inventory') & (groceries.Subcategory == 'Snacks') # filtering with multiple conditions

groceries[mask]
groceries.loc[mask, ['Subcategory','Item','Inventory']]
```

### Sorting 

```python
groceries.sort_values('Price_Dollars')
groceries.sort_values('Price_Dollars', ascending=True)
groceries.sort_values('Price_Dollars', ascending=False).head()
```

```python
groceries.sort_values(['Subcategory','Price_Dollars']).head()
```

```python
groceries[groceries['Low Inventory'] == 'Low Inventory'].sort_values('Price_Dollars')
```

### Grouping

```python
groceries.head()
```

```python
groceries.groupby('Category')['Inventory'].sum()
groceries.groupby('Category')['Inventory'].count()
```

```python
groceries.groupby(['Category','Subcategory'])['Inventory'].count()
groceries.groupby(['Category','Subcategory'])['Inventory'].count().reset_index()
```

```python
(groceries.groupby(['Category','Subcategory'])['Inventory']
          .agg(['sum','count'])
          .reset_index())
```

```python
groceries.groupby('Category')['Price_Dollars'].max()
```

```python
(groceries[['Category','Subcategory','Price_Dollars']]
    .sort_values('Price_Dollars', ascending=False)
    .groupby('Category').head(1))
```
