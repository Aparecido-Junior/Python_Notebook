### Handling inconsistent text & Typos

it is possible to fix inconsistnt text & typos by:

1. .loc[]
2. np.where() `np.where(condition, if_true, if_false)`
3. .map()
4. String methods

```python
# np.where()
# update values based on a logical condition

# Checking categories
df.Class.value_counts()

# Changing EDA category to Exploratory Data Analysis
df.Class = np.where(df.Class == 'EDA', 'Exploratory Data Analysis', df.Class)

# Cheking chaging
df.Class.value_counts()
```

```python
# Changing EDA category to Exploratory Data Analysis
df.Class = np.where(df.Class == 'Python', 'Intro to Python', df.Class)

# Cheking chaging
df.Class.value_counts()
```

```python
# Checking grade values
df.Grade.describe()

# Changing grade over 100
df.Grade = np.where(df.Grade > 100, 100, df.Grade) 

# # Checking changings
df.Grade.describe()
```

### Map values

Use .map() to map values from one set of values to another set of values.

### Duplicate Data

Duplicate data represents the presence of one or more redudant rows that contain the same information as another, and can therefore be removed.
